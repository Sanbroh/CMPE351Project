### RQ1: What are the dominant sentiment trends in financial news during periods of high versus low market volatility, and how do they correlate with sector-specific performance?

```python
# Import news data
df_news = pd.read_csv("nasdaq_exteral_data.csv")
print(df_news.head())

# Import screener
# Desired sectors: Tech, Financials, Basic Materials
df_screener = pd.read_csv("screener_Tech-Finance-BasicMaterials.csv")
print(df_screener.head())

# Import sample price data for view
df_price_test = pd.read_csv("full_history/A.csv")
print(df_price_test.head())
```

```python
# Process datasets & filter for desired tickers in the focus sectors

# Screener & filtering
screener_tickers = df_screener['Symbol'].unique()
df_news = df_news[df_news['Stock_symbol'].isin(screener_tickers)]
df_news['Date'] = pd.to_datetime(df_news['Date'])
print(df_news.head())

# Get data stats
print("Len:", len(df_news))
print("Filter_SIZE:", len(screener_tickers))
print("Ticker_SIZE:", df_news['Stock_symbol'].nunique())
```

```python
# Sampling
df_news_processed = df_news.sample(frac=1, random_state=42)
df_news_processed = df_news_processed.dropna(subset=['Stock_symbol'])
print(df_news_processed.head())
print("Len_Sample:", len(df_news_processed))
```

```python
def get_combined_summary(row):
    summaries = []
    for col in ['Lsa_summary', 'Luhn_summary', 'Textrank_summary', 'Lexrank_summary']:
        if pd.notnull(row[col]) and row[col].strip():
            summaries.append(row[col].strip())
    if summaries:
        # Combine all available summaries
        return " ".join(summaries)
    elif pd.notnull(row['Article']) and row['Article'].strip():
        # Fallback to the full article if no summaries exist
        return row['Article'].strip()
    else:
        return None

# Create the combined summary column for sentiment analysis
df_news_processed['text_for_sentiment'] = df_news_processed.progress_apply(get_combined_summary, axis=1)

# Drop rows without any valid summary text
df_news_processed = df_news_processed.dropna(subset=['text_for_sentiment'])
print("Number of rows after dropping missing summaries:", len(df_news_processed))

def is_english(text):
    try:
        return detect(text) == 'en'
    except Exception:
        return False

df_news_processed = df_news_processed[df_news_processed['text_for_sentiment'].progress_apply(is_english)]
print("Number of rows after dropping non-English texts:", len(df_news_processed))
```

```python
# Load the FinBERT sentiment-analysis pipeline
classifier = pipeline("sentiment-analysis", model="yiyanghkust/finbert-tone", tokenizer="yiyanghkust/finbert-tone")

# Prepare texts for sentiment analysis
texts = df_news_processed['text_for_sentiment'].tolist()
texts = [text[:512] for text in texts]

batch_size = 16
batched_texts = [texts[i:i+batch_size] for i in range(0, len(texts), batch_size)]
all_results = []
for batch in tqdm(batched_texts, desc="Processing sentiment batches"):
    batch_results = classifier(batch, truncation=True)
    all_results.extend(batch_results)

def parse_sentiment(result):
    """
    Convert FinBERT output to a sentiment score.
    Positive sentiment yields positive,
    Negative sentiment yields negative,
    Neutral sentiment returns 0.
    """
    label = result['label'].lower() if result else None
    score = result['score'] if result else 0
    if label == 'positive':
        return score
    elif label == 'negative':
        return -score
    else:
        return 0

# Generate sentiment scores
sentiment_scores = [parse_sentiment(r) for r in all_results]
df_news_processed['sentiment_score'] = sentiment_scores
print(df_news_processed[['Stock_symbol', 'sentiment_score']].head())
```

```python
df_news_processed[['Date', 'Stock_symbol', 'sentiment_score']].to_pickle('df_news_processed.pkl')
print(df_news_processed[['Date', 'Stock_symbol', 'sentiment_score']])
```

```python
print("Stage 1: Processing Financial News Data")
df_news_processed['Date'] = pd.to_datetime(df_news_processed['Date'])
df_news_processed['news_date'] = df_news_processed['Date'].dt.date

print("Stage 2: Loading and Processing Screener Data")
df_screener = pd.read_csv("screener_Tech-Finance-BasicMaterials.csv")
df_screener['Symbol'] = df_screener['Symbol'].str.strip()

# Merge the news data with screener to attach the Sector
news_data = df_news_processed.merge(df_screener[['Symbol', 'Sector']], left_on='Stock_symbol', right_on='Symbol', how='left')
news_data.drop(columns=['Symbol'], inplace=True)

print("Stage 3: Loading and Processing Market Data")
csv_files = glob.glob("full_history/*.csv")
market_dfs = []

for file in csv_files:
    df_temp = pd.read_csv(file)
    ticker = os.path.basename(file).split('.')[0]
    df_temp['ticker'] = ticker
    market_dfs.append(df_temp)

# Combine all
df_market = pd.concat(market_dfs, ignore_index=True)
df_market['date'] = pd.to_datetime(df_market['date']).dt.date
df_market['volatility'] = (df_market['high'] - df_market['low']) / df_market['close']
df_market = df_market.sort_values(by=['ticker', 'date'])
df_market['daily_return'] = df_market.groupby('ticker')['close'].pct_change()
df_market = df_market.merge(df_screener[['Symbol', 'Sector']], left_on='ticker', right_on='Symbol', how='left')
df_market.drop(columns=['Symbol'], inplace=True)

print("Stage 4: Aggregating Market Data by Sector and Overall")
market_sector = df_market.groupby(['date', 'Sector']).agg({
    'volatility': 'mean',
    'daily_return': 'mean'
}).reset_index().rename(columns={'daily_return': 'avg_return'})

market_overall = df_market.groupby('date').agg({
    'volatility': 'mean'
}).reset_index().rename(columns={'volatility': 'overall_volatility'})

market_sector = market_sector.merge(market_overall, on='date', how='left')
vol_median = market_overall['overall_volatility'].median()
market_sector['volatility_category'] = np.where(market_sector['overall_volatility'] >= vol_median, 'High', 'Low')

print("Stage 5: Aggregating News Sentiment by Date and Sector")
news_sector = news_data.groupby(['news_date', 'Sector']).agg({
    'sentiment_score': ['mean', 'count']
}).reset_index()
news_sector.columns = ['date', 'Sector', 'avg_sentiment', 'news_count']

# Merge on date and sector
print("Stage 6: Merging News Sentiment with Market Data")
merged_data = pd.merge(news_sector, market_sector, on=['date', 'Sector'], how='inner')

print("DONE")
```

```python
print("Stage 7: Analyzing Correlations Between Sentiment and Sector Performance")

# Compute the correlation
corr_results = (merged_data.groupby(['Sector', 'volatility_category'])
                .apply(lambda x: x['avg_sentiment'].corr(x['avg_return']))
                .reset_index())
corr_results.columns = ['Sector', 'Volatility_Category', 'sentiment_return_corr']

print("\nCorrelation between average sentiment and sector performance (average return):")
print(corr_results)

# Compute summary
summary_stats = merged_data.groupby(['Sector', 'volatility_category']).agg({
    'avg_sentiment': ['mean', 'median', 'std', 'min', 'max'],
    'avg_return': ['mean', 'median', 'std', 'min', 'max'],
    'news_count': 'sum'
}).reset_index()

summary_stats.columns = [
    'Sector', 'Volatility_Category', 
    'Sentiment_Mean', 'Sentiment_Median', 'Sentiment_Std', 'Sentiment_Min', 'Sentiment_Max',
    'Return_Mean', 'Return_Median', 'Return_Std', 'Return_Min', 'Return_Max',
    'Total_News_Count'
]

print("\nSummary statistics by Sector and Volatility Category:")
print(summary_stats)

# Textual summary
print("\nAnalysis Summary:")
for sector in summary_stats['Sector'].unique():
    for vol_cat in ['High', 'Low']:
        subset = summary_stats[(summary_stats['Sector'] == sector) & (summary_stats['Volatility_Category'] == vol_cat)]
        if not subset.empty:
            s_mean = subset['Sentiment_Mean'].values[0]
            r_mean = subset['Return_Mean'].values[0]
            news_count = subset['Total_News_Count'].values[0]
            corr_val = corr_results[(corr_results['Sector'] == sector) & (corr_results['Volatility_Category'] == vol_cat)]['sentiment_return_corr'].values[0]
            print(f"\nFor sector '{sector}' during {vol_cat} volatility periods:")
            print(f"  - Average sentiment: {s_mean:.3f}")
            print(f"  - Average abnormal return: {r_mean:.3f}")
            print(f"  - Total number of news items: {news_count}")
            print(f"  - Correlation between sentiment and abnormal return: {corr_val:.3f}")

print("DONE")
```

```python
# Graph 1-A: Average News Sentiment Over Time for High Volatility Days
print("Generating Graph 1-A: Average News Sentiment Over Time for High Volatility Days")
news_overall = news_data.groupby('news_date').agg({
    'sentiment_score': 'mean'
}).reset_index().rename(columns={'news_date': 'date', 'sentiment_score': 'avg_sentiment'})
news_overall = pd.merge(news_overall, market_overall, on='date', how='left')
news_overall['volatility_category'] = np.where(news_overall['overall_volatility'] >= vol_median, 'High', 'Low')
high_vol = news_overall[news_overall['volatility_category'] == 'High'].sort_values(by='date')

plt.figure(figsize=(10, 6))
plt.plot(high_vol['date'].to_numpy(), high_vol['avg_sentiment'].to_numpy(), marker='o')
plt.title("Average News Sentiment Over Time - High Volatility Days")
plt.xlabel("Date")
plt.ylabel("Average Sentiment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Graph 1-B: Average News Sentiment Over Time for Low Volatility Days
print("Generating Graph 1-B: Average News Sentiment Over Time for Low Volatility Days")
low_vol = news_overall[news_overall['volatility_category'] == 'Low'].sort_values(by='date')

plt.figure(figsize=(10, 6))
plt.plot(low_vol['date'].to_numpy(), low_vol['avg_sentiment'].to_numpy(), marker='o', color='green')
plt.title("Average News Sentiment Over Time - Low Volatility Days")
plt.xlabel("Date")
plt.ylabel("Average Sentiment")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### RQ2: To what extent does the sentiment trends of news sources predict abnormal trading volume and price changes for individual stocks?

```python
news_agg = news_data.groupby(['Stock_symbol', 'news_date']).agg({
    'sentiment_score': 'mean'
}).reset_index().rename(columns={'Stock_symbol': 'ticker',
                                'news_date': 'date',
                                'sentiment_score': 'avg_sentiment'})
print("News sentiment aggregated by ticker and date:", news_agg.shape)

df_market = df_market.sort_values(by=['ticker', 'date'])
window = 20

# Abnormal Trading Volume
df_market['vol_rolling_mean'] = df_market.groupby('ticker')['volume'].transform(lambda x: x.rolling(window=window, min_periods=5).mean())
df_market['vol_rolling_std'] = df_market.groupby('ticker')['volume'].transform(lambda x: x.rolling(window=window, min_periods=5).std())
df_market['abnormal_volume'] = (df_market['volume'] - df_market['vol_rolling_mean']) / df_market['vol_rolling_std']

# Abnormal Price Change
df_market['ret_rolling_mean'] = df_market.groupby('ticker')['daily_return'].transform(lambda x: x.rolling(window=window, min_periods=5).mean())
df_market['ret_rolling_std'] = df_market.groupby('ticker')['daily_return'].transform(lambda x: x.rolling(window=window, min_periods=5).std())
df_market['abnormal_return'] = (df_market['daily_return'] - df_market['ret_rolling_mean']) / df_market['ret_rolling_std']

print("Market data with abnormal volume and return computed:", df_market.shape)

# Merge
merged = pd.merge(df_market, news_agg, on=['ticker', 'date'], how='inner')
print("Merged dataset shape (market + news):", merged.shape)

# Calc correlations
corr_vol = merged['avg_sentiment'].corr(merged['abnormal_volume'])
corr_ret = merged['avg_sentiment'].corr(merged['abnormal_return'])
print("Correlation between avg_sentiment and abnormal_volume:", corr_vol)
print("Correlation between avg_sentiment and abnormal_return:", corr_ret)

# Regression for abnormal volume
X_vol = sm.add_constant(merged['avg_sentiment'])
model_volume = sm.OLS(merged['abnormal_volume'], X_vol, missing='drop').fit()
print("Regression results for abnormal volume:\n", model_volume.summary())

# Regression for abnormal return
X_ret = sm.add_constant(merged['avg_sentiment'])
model_return = sm.OLS(merged['abnormal_return'], X_ret, missing='drop').fit()
print("Regression results for abnormal return:\n", model_return.summary())
```

```python
# Graph A: Hexbin Plot for Abnormal Return
fig, ax = plt.subplots(figsize=(10, 6))
hb = ax.hexbin(merged['avg_sentiment'].to_numpy(), merged['abnormal_return'].to_numpy(), 
               gridsize=50, cmap='Blues', mincnt=1)
ax.set_title("Hexbin: News Sentiment vs. Abnormal Return")
ax.set_xlabel("Average Sentiment")
ax.set_ylabel("Abnormal Return (z-score)")
cb = plt.colorbar(hb, ax=ax)
cb.set_label("Counts")
plt.tight_layout()
plt.show()

# Graph B: Hexbin Plot for Abnormal Volume
fig, ax = plt.subplots(figsize=(10, 6))
hb = ax.hexbin(merged['avg_sentiment'].to_numpy(), merged['abnormal_volume'].to_numpy(), 
               gridsize=50, cmap='Blues', mincnt=1)
ax.set_title("Hexbin: News Sentiment vs. Abnormal Volume")
ax.set_xlabel("Average Sentiment")
ax.set_ylabel("Abnormal Volume (z-score)")
cb = plt.colorbar(hb, ax=ax)
cb.set_label("Counts")
plt.tight_layout()
plt.show()

# Graph C: Histograms of Abnormal Return by Sentiment Group
median_sentiment = merged['avg_sentiment'].median()
high_sentiment_data = merged[merged['avg_sentiment'] >= median_sentiment]['abnormal_return']
low_sentiment_data = merged[merged['avg_sentiment'] < median_sentiment]['abnormal_return']

fig, ax = plt.subplots(figsize=(10, 6))
bins = np.linspace(merged['abnormal_return'].min(), merged['abnormal_return'].max(), 50)
ax.hist(high_sentiment_data, bins=bins, alpha=0.6, label='High Sentiment', density=True)
ax.hist(low_sentiment_data, bins=bins, alpha=0.6, label='Low Sentiment', density=True)
ax.set_title("Distribution of Abnormal Return by Sentiment Group")
ax.set_xlabel("Abnormal Return (z-score)")
ax.set_ylabel("Density")
ax.legend()
plt.tight_layout()
plt.show()

# Graph D: Rolling Correlation between Daily Avg Sentiment & Abnormal Return
daily_data = merged.groupby('date').agg({
    'avg_sentiment': 'mean',
    'abnormal_return': 'mean'
}).reset_index()
daily_data = daily_data.sort_values('date')
rolling_corr = daily_data['avg_sentiment'].rolling(window=30, min_periods=10).corr(daily_data['abnormal_return'])

fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(daily_data['date'].to_numpy(), rolling_corr.to_numpy(), marker='o', linestyle='-')
ax.set_title("30-Day Rolling Correlation: Average Sentiment vs. Abnormal Return")
ax.set_xlabel("Date")
ax.set_ylabel("Rolling Correlation")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### RQ3: How effectively can sentiment-enhanced factor models outperform traditional minimum volatility models in predicting risk-adjusted returns

```python
# Load from RQ1
df = df_news_processed
df['Date'] = pd.to_datetime(df['Date']).dt.date

print(df)

csv_cache = {}
closing_prices = []
volumes = []
folder_path = 'full_history'

for idx, row in tqdm(df.iterrows(), total=len(df), desc="Processing rows"):
    symbol = row['Stock_symbol']
    date = row['Date']

    if symbol not in csv_cache:
        file_path = os.path.join(folder_path, f"{symbol}.csv")
        if os.path.exists(file_path):
            stock_df = pd.read_csv(file_path, parse_dates=['date'])
            stock_df['date'] = stock_df['date'].dt.date
            csv_cache[symbol] = stock_df
        else:
            csv_cache[symbol] = None

    stock_data = csv_cache[symbol]

    if stock_data is not None:
        match = stock_data[stock_data['date'] == date]
        if not match.empty:
            closing_prices.append(match.iloc[0]['close'])
            volumes.append(match.iloc[0]['volume'])
        else:
            closing_prices.append(None)
            volumes.append(None)
    else:
        closing_prices.append(None)
        volumes.append(None)

df['close'] = closing_prices
df['volume'] = volumes

print(df)
```

```python
df = pd.read_pickle("trading_universe.pkl")

df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['month_year'] = df['Date'].dt.to_period('M')

# Select least volatile
least_vol_stocks = []

for period in df['month_year'].unique():
    monthly_data = df[df['month_year'] == period]
    last_per_stock = monthly_data.sort_values('Date').groupby('Stock_symbol').tail(1)
    
    n = len(last_per_stock)
    decile_count = max(1, int(np.floor(n * 0.10)))  # Ensure at least one stock is selected
    top_decile = last_per_stock.sort_values(by='rolling_volatility').head(decile_count)
    
    least_vol_stocks.append(top_decile)

# Combine
least_vol_stocks = pd.concat(least_vol_stocks).reset_index(drop=True)
least_vol_stocks['weight'] = 1 / least_vol_stocks.groupby('month_year')['Stock_symbol'].transform('count')

print(f"Shape of the least volatile stocks selection: {least_vol_stocks.shape}")
print("First few rows of the selected least volatile stocks:")
print(least_vol_stocks[['month_year', 'Stock_symbol', 'rolling_volatility', 'weight', 'close']].head(11))

least_vol_stocks.to_pickle("low_vol_portfolio.pkl")
```

```python
least_vol_stocks = pd.read_pickle("low_vol_portfolio.pkl")

df = pd.read_pickle("trading_universe.pkl")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['month_year'] = df['Date'].dt.to_period('M')

month_end_prices = (
    df.sort_values('Date')
    .groupby(['Stock_symbol', 'month_year'])
    .tail(1)
    .copy()
)
month_end_prices['prev_close'] = month_end_prices.groupby('Stock_symbol')['close'].shift(1)
month_end_prices['monthly_return'] = (month_end_prices['close'] / month_end_prices['prev_close']) - 1

least_vol_stocks['next_month'] = least_vol_stocks['month_year'] + 1
returns_df = month_end_prices[['Stock_symbol', 'month_year', 'monthly_return']]
returns_df.columns = ['Stock_symbol', 'next_month', 'monthly_return']

portfolio_with_returns = pd.merge(
    least_vol_stocks,
    returns_df,
    on=['Stock_symbol', 'next_month'],
    how='left'
)
portfolio_with_returns['weighted_return'] = (
    portfolio_with_returns['weight'] * portfolio_with_returns['monthly_return']
)
portfolio_returns = (
    portfolio_with_returns.groupby('month_year')['weighted_return']
    .sum()
    .reset_index()
    .rename(columns={'weighted_return': 'portfolio_return'})
)

print("Monthly portfolio returns:")
print(portfolio_returns.head(100))

portfolio_returns.to_pickle("low_vol_portfolio_returns.pkl")
```

```python
least_vol_stocks = pd.read_pickle("low_vol_positive_sentiment_portfolio.pkl")

df = pd.read_pickle("trading_universe.pkl")
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['month_year'] = df['Date'].dt.to_period('M')

month_end_prices = (
    df.sort_values('Date')
    .groupby(['Stock_symbol', 'month_year'])
    .tail(1)
    .copy()
)
month_end_prices['prev_close'] = month_end_prices.groupby('Stock_symbol')['close'].shift(1)
month_end_prices['monthly_return'] = (month_end_prices['close'] / month_end_prices['prev_close']) - 1

least_vol_stocks['next_month'] = least_vol_stocks['month_year'] + 1
returns_df = month_end_prices[['Stock_symbol', 'month_year', 'monthly_return']]
returns_df.columns = ['Stock_symbol', 'next_month', 'monthly_return']

portfolio_with_returns = pd.merge(
    least_vol_stocks,
    returns_df,
    on=['Stock_symbol', 'next_month'],
    how='left'
)
portfolio_with_returns['weighted_return'] = (
    portfolio_with_returns['weight'] * portfolio_with_returns['monthly_return']
)
portfolio_returns = (
    portfolio_with_returns.groupby('month_year')['weighted_return']
    .sum()
    .reset_index()
    .rename(columns={'weighted_return': 'portfolio_return'})
)

print("Monthly portfolio returns:")
print(portfolio_returns.head(100))

portfolio_returns.to_pickle("low_vol_sentiment_portfolio_returns.pkl")
```

```python
# Load both sets of monthly portfolio
low_vol = pd.read_pickle("low_vol_portfolio_returns.pkl").copy()
sentiment_vol = pd.read_pickle("low_vol_sentiment_portfolio_returns.pkl").copy()

low_vol['date'] = low_vol['month_year'].dt.to_timestamp()
sentiment_vol['date'] = sentiment_vol['month_year'].dt.to_timestamp()

low_vol = low_vol.rename(columns={'portfolio_return': 'low_vol_return'})
sentiment_vol = sentiment_vol.rename(columns={'portfolio_return': 'sentiment_return'})

combined = pd.merge(low_vol[['date', 'low_vol_return']],
                    sentiment_vol[['date', 'sentiment_return']],
                    on='date',
                    how='inner')

combined['low_vol_cum'] = (1 + combined['low_vol_return']).cumprod()
combined['sentiment_cum'] = (1 + combined['sentiment_return']).cumprod()

plt.figure(figsize=(12, 6))
plt.plot(combined['date'].values, combined['low_vol_cum'].values, label="Low Vol Portfolio", linewidth=2)
plt.plot(combined['date'].values, combined['sentiment_cum'].values, label="Low Vol + Sentiment Portfolio", linewidth=2)
plt.title("Cumulative Returns Comparison")
plt.xlabel("Date")
plt.ylabel("Cumulative Return")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

def annualized_sharpe(returns):
    return (returns.mean() / returns.std()) * np.sqrt(12)

def max_drawdown(cum_returns):
    roll_max = cum_returns.cummax()
    drawdown = (cum_returns - roll_max) / roll_max
    return drawdown.min()

def CAGR(cum_returns, dates):
    total_years = (dates.iloc[-1] - dates.iloc[0]).days / 365.25
    return cum_returns.iloc[-1]**(1 / total_years) - 1

def calmar_ratio(cagr, mdd):
    return cagr / abs(mdd) if mdd != 0 else np.nan

sharpe_low_vol = annualized_sharpe(combined['low_vol_return'])
sharpe_sentiment = annualized_sharpe(combined['sentiment_return'])

mdd_low_vol = max_drawdown(combined['low_vol_cum'])
mdd_sentiment = max_drawdown(combined['sentiment_cum'])

cagr_low_vol = CAGR(combined['low_vol_cum'], combined['date'])
cagr_sentiment = CAGR(combined['sentiment_cum'], combined['date'])

calmar_low_vol = calmar_ratio(cagr_low_vol, mdd_low_vol)
calmar_sentiment = calmar_ratio(cagr_sentiment, mdd_sentiment)

summary_df = pd.DataFrame({
    'CAGR': [cagr_low_vol, cagr_sentiment],
    'Sharpe Ratio': [sharpe_low_vol, sharpe_sentiment],
    'Max Drawdown': [mdd_low_vol, mdd_sentiment],
    'Calmar Ratio': [calmar_low_vol, calmar_sentiment]
}, index=['Low Vol', 'Low Vol + Sentiment'])

print(" Strategy Performance Summary:")
formatted_df = summary_df.copy()
formatted_df["CAGR"] = formatted_df["CAGR"].map("{:.4%}".format)
formatted_df["Sharpe Ratio"] = formatted_df["Sharpe Ratio"].map("{:.4f}".format)
formatted_df["Max Drawdown"] = formatted_df["Max Drawdown"].map("{:.2%}".format)
formatted_df["Calmar Ratio"] = formatted_df["Calmar Ratio"].map("{:.4f}".format)
print(formatted_df)

drawdown_lv = combined['low_vol_cum'] / combined['low_vol_cum'].cummax() - 1
drawdown_sent = combined['sentiment_cum'] / combined['sentiment_cum'].cummax() - 1

plt.figure(figsize=(12, 4))
plt.plot(combined['date'].values, drawdown_lv.values, label="Low Vol Drawdown", linestyle='--')
plt.plot(combined['date'].values, drawdown_sent.values, label="Sentiment-Enhanced Drawdown", linestyle='--')
plt.title("Drawdown Over Time")
plt.xlabel("Date")
plt.ylabel("Drawdown")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
